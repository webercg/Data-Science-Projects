```python
import pandas as pd
```

```python
df = pd.read_csv('dados_tratados.csv', delimiter=',', encoding='utf-8')
df
```

# Tratamentos

## Nome de Colunas

```python
df.columns = [x.strip() for x in df.columns]
```

## Tipo colunas

```python
df['Charm'] = df['Charm'].astype(int)
df['Exp'] = df['Exp'].astype(int)
df['HP'] = df['HP'].astype(int)
```

# Metricas

## Charm por HP total * kill

```python
df['HP_Total_Desbloqueio'] = df['HP'] * df['Kills']
df.head()
```

```python
df['Charm_por_HP_removido_creatura'] = df['Charm'] / df['HP_Total_Desbloqueio']
df
```

# Insights

## Quais são as principais dificuldades das criaturas?

```python
df.Dificuldade.value_counts()
```

## Quais são as ocorrencias das criaturas?

```python
df.Ocorrencia.value_counts()
```

## Até quantos pontos cada criatura pode dar?

```python
df.Charm.value_counts()
```

## Como a quantidade de criaturas se distribui em termos da ocorrencia, dificuldade e os pontos que cada uma podem conceder de charm?

```python
df_aux = df.groupby(['Ocorrencia','Dificuldade','Charm']).agg({'Criatura':"count", 'HP':'sum'}).reset_index().rename(columns={'Criatura':'Qtd_criaturas', 'HP':'HP_total'})
df_aux['Charms_totais'] = df_aux['Qtd_criaturas']*df_aux['Charm']
df_aux['Charms_por_HP_total'] = df_aux['Charm']/df_aux['HP_total']
df_aux
```

```python
df_aux.sort_values('Charms_totais', ascending=False)
```

- Criaturas comuns, hard, medium, challeging e easy são as que fornecem maiores quantidade de charms

```python
df_aux.sort_values('Charms_por_HP_total', ascending=False)
```

- Criaturas que dão mais charms por hp total tem ocorrencias extremamente raras

```python
df_aux.sort_values('Qtd_criaturas', ascending=False)
```

- Criaturas de ocorrencia comum e dificuldades hard, medium e easy são em maior qtde e há maior opção de fazer o bestiary

## Quais são as criaturas comuns de dificuldades hard medium e easy que dão mais charm por hp total?

```python
df_aux = df[ (df['Dificuldade'].isin(['Medium','Hard','Easy'])) & (df['Ocorrencia']=='Common')].sort_values(['Charm_por_HP_removido_creatura','Kills','Charm'], ascending=[False,False,False])
df_aux
```
