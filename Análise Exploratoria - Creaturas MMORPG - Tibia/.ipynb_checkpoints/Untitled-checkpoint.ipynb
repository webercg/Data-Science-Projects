```python
import pandas as pd
```

```python
df = pd.read_csv('dados_tratados.csv', delimiter=',', encoding='utf-8')
df
```

## Tratamentos

```python
df['Charm'] = df['Charm'].apply(lambda x: x.replace(' Charm points',''))
df['Exp'] = df['Exp'].apply(lambda x: x.replace(' Experience',''))
df['HP'] = df['HP'].apply(lambda x: x.replace(' Hitpoints',''))
df.head()
```

```python
df['Charm'] = df['Charm'].astype(int)
df['Exp'] = df['Exp'].astype(int)
df['HP'] = df['HP'].astype(int)
```

## Metricas

```python
df['HP_Total_Desbloqueio'] = df['HP'] * df['Kills']
df.head()
```

```python
df['Charm_por_HP_removido_creatura'] = df['Charm'] / df['HP_Total_Desbloqueio']
df
```

```python
df.sort_values('Charm_por_HP_removido_creatura', ascending=False)
```
