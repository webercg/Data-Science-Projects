### Solicitando requisições ao spotify e coletando audio_features

```python
lista_features = []
tracks_erros = []

for musica in tracks_unicas:
    #captando dados
    #print(musica)
    resultado = sp.audio_features(tracks=[musica])
    
    #transformando para dicionário
    dicionario = resultado[0]
    
    # Se não retornar nenhum valor guardar na variável de musicas com erros
    if dicionario is None:
        tracks_erros.append(musica)
        
    else:
    
        #adicionando URL:
        #dicionario['URL'] = musica
    
        #enumerando as keys e values do dicionário
        lista_chaves = list(dicionario.keys())
        lista_values = list(dicionario.values())
    
        #Guardando a informação da URL da musica
        lista_chaves.append('URL')
        lista_values.append(musica)
    
        #Guardando toda informação minerada
        lista_features.append(np.array(lista_values))
    
    
    
        #Formatando o dicionário
        #my_dict = {}
        #for i, chave in enumerate(lista_chaves):
            #my_dict[chave] = [lista_values[i]]
    
        #transformando para dataframe
        #df_aux = pd.DataFrame.from_dict(my_dict)
    
        #concatenando resultados
        #df_audio_features = pd.concat([df_audio_features, df_aux])
```

```python
#transformando os resultados para dataframe:
df_features = pd.DataFrame(np.array(lista_features), columns=lista_chaves)
df_features
```
