## Capturando audiofeatures + popularidade do cantor

```python
resultados = []
lista_values = []
tracks_erros = []

for i,musica in enumerate(tracks_unicas):
    #captando dados
    resultado = sp.audio_features(tracks=[musica])
    print(i)
    #transformando para dicionário
    dicionario = resultado[0]
    
    # Se não retornar nenhum valor guardar na variável de musicas com erros
    if dicionario is None:
        tracks_erros.append(musica)
        
    else:
    
    
        #values do dicionário
        lista_values = list(dicionario.values())
    

        # adicionando track id a lista
        lista_values.append(musica)
    
    
        # procurando pela popularidade, seguidores e genero do artista
        ## Procurando o artista da musica
        faixa = sp.track(musica)
        faixa

        ## Procurando pelo artista
        artista = faixa['artists'][0]['id']
        artista_valores = sp.artist(artista)
        popularidade = artista_valores['popularity']
        seguidores = artista_valores['followers']['total']
        genero = ', '.join(artista_valores['genres'])
        
        if len(artista_valores['genres'])>0:
            genero_primario = artista_valores['genres'][0]
        else:
            genero_primario = 'NaN'

        
        
        #Guardando toda informação minerada
        lista_values.append(popularidade)
        lista_values.append(seguidores)
        lista_values.append(genero)
        lista_values.append(genero_primario)
    
        
        #Armazenando em formato np.array para transformar em dataframe
        resultados.append(np.array(lista_values))

#Transformando em DF
```

```python
lista_chaves = list(dicionario.keys()) + ['URL','popularidade_cantor','seguidores_cantor','genero_cantor','genero_primario_cantor']
lista_chaves
```

```python
resultados_df = pd.DataFrame(np.array(resultados), columns=lista_chaves)
resultados_df
```

```python
resultados_df.to_csv('audio_features2.csv')
```
