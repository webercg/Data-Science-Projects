# 3 Processamento e Transformações

## 3.1 Vencedor da partida: mandante ou visitante

```python
## Vencedor mandante ou visitante
condicoes = [(df_campeonato['vencedor']==df_campeonato['mandante']),\
             (df_campeonato['vencedor']==df_campeonato['visitante']),\
             (df_campeonato['vencedor']=='-')]

resultado_condicoes = ['mandante','visitante','empate']
df_campeonato['vencedor_mandante_visitante'] = np.select(condicoes,resultado_condicoes,default=np.nan)

df_campeonato.head()
```

## 3.2 Diferença de gols entre os clubes na partida

```python
## Diferença de gols entre os clubes na partida
df_campeonato['diff_gols'] = abs(df_campeonato['mandante_Placar']-df_campeonato['visitante_Placar'])
df_campeonato.head()
```

## 3.3 Dias de desanso desde o ultimo jogo - mandante e visitante

```python
df_campeonato.mandante.unique()
```

```python
df_campeonato.visitante.unique()
```

```python
## Transformando coluna pra data
df_campeonato['data'] = pd.to_datetime(df_campeonato['data'])
```

```python
## Colocando os jogos em ordem cronológica
df_campeonato = df_campeonato.sort_values(by = 'data', ascending = True)
df_campeonato.head()
```

```python
df_dias_de_descanso = pd.DataFrame()

for time in df_campeonato.mandante.unique().tolist():
    print(time)
    
    df_time = df_campeonato.query('mandante == ' + '"' +  str(time) + '"'  + ' | visitante == ' + '"' + str(time) + '"' )
    
    
    df_aux = df_time[['data']].diff(axis=0, periods = 1).reset_index().rename(columns={'data':'dias_de_descanso'})
    df_aux['dias_de_descanso'] = df_aux['dias_de_descanso'].dt.days
    df_aux['time'] = time
    df_dias_de_descanso = pd.concat([df_dias_de_descanso, df_aux])  
```
